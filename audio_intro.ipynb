Загружаем библиотеки

````python
import scipy.io.wavfile as wavfile
import IPython.display as display
import matplotlib.pyplot as plt
````

Открываем аудио файл

wav - формат (как сжимать аудио)

riff - контейнер (как хранить блоки данных)

Пример через команду file:
```
file dataset/wav/6_9_1_9.wav

RIFF (little-endian) data, WAVE audio, Microsoft PCM, 16 bit, mono 48000 Hz
```

16 bit - Precision - сколько памяти выделено на одно число из массива (значения $[-2^{15}: 2^{15} - 1] = [-32768:32767]$)

48000 Hz - количество элементов массива за 1 секунду времени

mono - 1 - количество каналов

### низкоуровневое издевательство
